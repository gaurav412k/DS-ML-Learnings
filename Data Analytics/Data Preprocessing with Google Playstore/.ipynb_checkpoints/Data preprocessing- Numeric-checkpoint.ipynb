```python
import pandas as pd
```

```python
data =pd.read_csv('googleplaystore.csv')
```

```python
data.head()
```

## Q1. Finding Average Rating

```python
data=data.dropna()
```

```python
ct=0
for i in data['Rating']:
    ct+=i
a= ct/len(data['Rating'])
print("Average of Rating is :",round(a,3))    
```

## Q2. How many apps are there with rating 5?

```python
ct=0
for i in data['Rating']:
    if (i==5.0):
        ct+=1

print("No. of apps with rating 5 is :",ct)    
```

## Q3. How many apps are there with rating between 4 - 4.5?

```python
ct=0
for i in data['Rating']:
    if (i>=4.0 and i<=4.5):
        ct+=1

print("No. of apps with rating 5 is :",ct) 
```

## Q4. Average App Reviews

```python
ct=0
for i in data['Reviews']:
    ct+=int(i)
a= ct/len(data['Reviews'])
print("Average of Reviews is :",round(a))  
```
