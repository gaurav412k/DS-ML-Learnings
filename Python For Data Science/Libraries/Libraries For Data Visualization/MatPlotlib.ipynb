```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
x = np.linspace(0, 4*np.pi, 100) 
#define 
sin_y = np.sin(x)
cos_y = np.cos(x)


plt.plot(x , sin_y, label="Sine")
plt.plot(x , cos_y, label="Cosine")
plt.title("sin and cosine wave")
plt.style.use('classic')
plt.legend(loc=1)

plt.show()

# legend()
# plot(x,y,label='')
# title()
# np.linspace(1,3.14,10)
# '*', '<'
# show()
# plt.style.use('_classic_test_patch')
```

```python
#Different types of plot
x =np.linspace(0,4*np.pi,100)
y = np.sin(x)

plt.style.use('Solarize_Light2')
#plot 
plt.plot(x,y, color= 'Red', linewidth=4, label='Sine' )

plt.legend()
plt.title('Sin wave')

plt.show()
```

```python
#Multiple line plots

x =np.linspace(0,4*np.pi,100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(8,2))
plt.style.use('Solarize_Light2')
#plot 
plt.plot(x,y_sin, color= 'Red', linewidth=4, label='Sine' )
plt.plot(x,y_cos, color= 'Blue', linewidth=4, label='Cosine' )

plt.legend(loc='upper right')
plt.title('Sin wave')

plt.axis('equal')
plt.show()
```

```python
cars = ['AUDI', 'BMW', 'TESLA', 'JAGUAR', 'MERCEDES']
data =  [ 450  , 385  ,    100  ,   200   ,    500    ]
explode = [0, 0.2, 0, 0.1, 0 ] 
fig = plt.figure(figsize =(7,7))


plt.pie(data, labels = cars, explode = explode, shadow = True , autopct = '%2.f%%' )

plt.show()
```

```python
#Adding weights
fig = plt.figure(figsize =(7,7))

plt.pie(data, labels = cars, explode = explode, shadow = True , autopct = '%2.f%%' ,
       wedgeprops = {'edgecolor' : 'black' , 'linewidth' : 2, 'antialiased' : True })

plt.show()
```

```python
# 4.6 Multiple Subplots 
x = np.linspace(0,10,50)

fig, ax = plt.subplots(2,2)

plt.style.use('seaborn-v0_8-dark')


y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('sin wave')

y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('Cos wave')

y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('Tan wave')

y = np.random.randn(1000)
ax[1,1].hist(y, bins = 40)
ax[1,1].set_title('Random Distribution')


plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0,10,50)
y = np.sin(x)

fig, (ax1,ax2,ax3) = plt. subplots(1,3, figsize = (18,6))
ax1.plot(x,y)
ax2.plot(x,y**2)
ax3.plot(x,y**3)

ax1.set_title('Sine wave')
ax2.set_title('Sine wave')
ax3.set_title('Sine wave')

plt.tight_layout()
plt.show()
```
